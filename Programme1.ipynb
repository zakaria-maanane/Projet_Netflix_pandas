```python
# Examiner d'abord les valeurs uniques dans la colonne
print(df['duration'].unique())

# Cr√©er une fonction pour extraire le nombre 
def extract_numeric(value):
    if isinstance(value, str):
        # Extraire seulement le nombre au d√©but de la cha√Æne
        import re
        match = re.search(r'^\d+', value)
        if match:
            return float(match.group())
    return value

# Appliquer la fonction √† la colonne
df['duration_numeric'] = df['duration'].apply(extract_numeric)

# V√©rifier le r√©sultat
print(df[['duration', 'duration_numeric']].head())
```

```python
import re

# Fonction pour extraire le nombre en d√©but de cha√Æne
def extract_number(value):
    if isinstance(value, str):
        match = re.search(r'^\d+', value)
        if match:
            return float(match.group())
    return value

# Appliquer cette fonction √† la colonne 'duration'
df['duration'] = df['duration'].apply(extract_number)

# V√©rifier le r√©sultat
print(df['duration'].dtypes)
print(df['duration'].head())
```
